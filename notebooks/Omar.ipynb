```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
csv1 = "../data/Physician_Compare_2015_Individual_EP_Public_Reporting.csv"
csv2 = "../data/Physician_Compare_2016_Individual_EP_Public_Reporting.csv"
csv3 = "../data/Physician_Compare_2015_Group_Patient_Experience.csv"
csv4 = "../data/Physician_Compare_2016_Group_Patient_Experience.csv"

output_file1 = "../Presentation/Clean/pcgroup_merged.csv"
output_file2 = "../Presentation/Clean/pcgroup_15.csv"
output_file3 = "../Presentation/Clean/pcgroup_16.csv"

pc_2015 = pd.read_csv(csv1)
pc_2016 = pd.read_csv(csv2)
pc_group_2015 = pd.read_csv(csv3)
pc_group_2016 = pd.read_csv(csv4)
```
