```python
group_file = "../Presentation/Clean/pcgroup_merged.csv"
group_15 = "../Presentation/Clean/pcgroup_15.csv"
group_16 = "../Presentation/Clean/pcgroup_16.csv"
opioid_15 = "../Presentation/Clean/op_final_15.csv"
opioid_16 = "../Presentation/Clean/op_final_16.csv"

output1 = '../Presentation/Clean/op_pac_final_15.csv'
output2 = '../Presentation/Clean/op_pac_final_16.csv'
output3 = '../Presentation/Clean/state_op_final_15.csv'
output4 = '../Presentation/Clean/state_op_final_16.csv'


pc_groupdf = pd.read_csv(group_file)
pc_group15df = pd.read_csv(group_15)
pc_group16df = pd.read_csv(group_16)
opioid_15df = pd.read_csv(opioid_15)
opioid_16df = pd.read_csv(opioid_16)
```

```python
#Create a new data frame that groups by PAC ID and sums up the Total Claim Count
opioid_15_pac = pd.DataFrame(opioid_15df.groupby(["PAC ID"])["Total Claim Count"].sum())
#Test
opioid_15_pac.head()
```

```python
#Use a groupby to sum the total opioid claim count per PAC ID#
opioid_15_pac["Opioid Claim Count"] = opioid_15df.groupby(["PAC ID"])["Opioid Claim Count"].sum()
#Test
opioid_15_pac.head()
#Create a column that comes up with percentage of of the opioid prescribing count
opioid_15_pac["Opioid Prescribing Rate"] = round((opioid_15_pac["Opioid Claim Count"] / opioid_15_pac["Total Claim Count"])*100,2)
#Test
opioid_15_pac.head()
```

```python
len(opioid_15_pac)
```

```python
# Create a new table for the PAC ID and States
pac_state15 = opioid_15df[["PAC ID", "State"]].copy()
pac_state15
#drop duplicates
pac_state15.drop_duplicates(keep=False)
pac_state15.head()
#opioid_15_pac.head()
```

```python
#combine our state table based on PAC ID to our opioid_pac 
opioid_15_pac1 = opioid_15_pac.merge(pac_state15, on= "PAC ID", how="outer")

opioid_15_pac1.head()
```

```python
#export the new file to our clean data folder
opioid_15_pac1.to_csv(output1, encoding="utf-8", index=False)
```

```python
#check the number of states in our analysis, we realize it includes territories as well
states15 = opioid_15_pac1["State"].unique()
print(states15)
print(len(states15))
```

```python
uniq = opioid_15_pac1["PAC ID"].nunique()
uniq
```

```python
#create a new df with information by state for 2015
state15_op_group = opioid_15_pac1.groupby(["State"])[["Total Claim Count","Opioid Claim Count"]].sum()
state15_op_group.head()
```

```python
#add the state opioid prescribing rate for 2016

state15_op_group["Opioid Prescribing Rate"] = round((state15_op_group["Opioid Claim Count"] / state15_op_group["Total Claim Count"])*100,2)
state15_op_group.head()
```

```python
#Sort by the highest Prescribing Rates
state_op_15final = state15_op_group.sort_values("Opioid Prescribing Rate", ascending=False)
state_op_15final.head()
```

```python
#export the state grouped file for 15 to our clean data folder
state_op_15final.to_csv(output3, encoding="utf-8")
```
